```python
# !/usr/bin/env python
#-*- coding:utf-8 -*-
# author: blue_dawn
# datetime:2019/5/28 3:59
# software: PyCharm
# function: 清晰地展现出出两个圆柱体相交的部分


# 导入模块
from  matplotlib import pyplot as plt 
%matplotlib
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
from matplotlib.font_manager import FontProperties
import numpy as np


# 处理中文显示问题
font = FontProperties(fname=r'C:\Windows\fonts\simsun.ttc', size=15)


# 定义图像窗口
fig = plt.figure(num= 1 , figsize= (8,5))


# 在窗口上添加3D坐标轴
ax = Axes3D(fig)
line, = ax.plot3D([],[],[],'r',linewidth=3.0)
line1, = ax.plot3D([],[],[], 'purple',linewidth=3.0)


# 两圆柱体的数据
u1 = np.linspace(0, 2 * np.pi, 200)
h1 = np.linspace(0, 1, 100)
h2 = np.linspace(-1, 1, 100)
x = np.outer(np.sin(u1), np.ones(len(h1)))
y1 = np.outer(np.cos(u1), np.ones(len(h1)))
y2 = np.outer(np.ones(len(u1)), h2)
z1 = np.outer(np.ones(len(u1)), h2)
z2 = np.outer(np.cos(u1), np.ones(len(h1)))

# 相交球的数据
u2, v2 = np.mgrid[0:2 * np.pi:200j, 0:np.pi:100j]
X = np.cos(u2) * np.sin(v2)
Y = np.sin(u2) * np.sin(v2)
Z = np.cos(v2)


def SetData_Animation(DataCount, Dims, DimLength, Kind):
    '''
    获取动态数据集函数
    '''
    DataCount_Recur = 2 * DataCount
    Data_Animation = np.empty((DataCount_Recur, Dims, DimLength))
    # 初始化起点
    if Kind == 1:
        u = np.linspace(0, 2 * np.pi, DimLength)
        x_Animation = np.cos(u)
        y_Animation = np.ones(len(u))
        z_Animation = np.sin(u)
        Data_Animation[0] = [x_Animation, y_Animation, z_Animation]
        # 步长
        Step = [np.zeros_like(x_Animation), (2 / (DataCount - 2)) * np.ones_like(y_Animation),
                np.zeros_like(z_Animation)]
        for index in range(1, DataCount_Recur):
            # 下一步的位置
            if index < DataCount-1:
                Data_Animation[index] = Data_Animation[index - 1] - Step
            elif (index == DataCount - 1) or (index == DataCount_Recur):
                Data_Animation[index] = Data_Animation[index - 1]
            else:
                Data_Animation[index] = Data_Animation[index - 1] + Step
        return Data_Animation
    elif Kind == 2:
        u = np.linspace(0, 2 * np.pi, DimLength)
        x_Animation = np.cos(u)
        y_Animation =  np.sin(u)
        z_Animation =np.ones(len(u))
        Data_Animation[0] = [x_Animation, y_Animation, z_Animation]
        # 步长
        Step = [np.zeros_like(x_Animation), np.zeros_like(z_Animation),
                (2 / (DataCount - 2)) * np.ones_like(y_Animation)]
        for index in range(1, DataCount_Recur):
            # 下一步的位置
            if index < DataCount - 1:
                Data_Animation[index] = Data_Animation[index - 1] - Step
            elif (index == DataCount - 1) or (index == DataCount_Recur):
                Data_Animation[index] = Data_Animation[index - 1]
            else:
                Data_Animation[index] = Data_Animation[index - 1] + Step
        return Data_Animation

    
def UpdateData_Animation(n):
    '''
    更新动态数据函数
    '''
    # 创建动态图形数据列表
    SetDataAnimation = SetData_Animation(40, 3, 50, 1)
    SetDataAnimation1 = SetData_Animation(40, 3, 50, 2)
    Recur = 2 * n
    for i in range(1, Recur):
        line.set_data(SetDataAnimation[i, 0:2])
        line.set_3d_properties(SetDataAnimation[i][2])
        line1.set_data(SetDataAnimation1[i, 0:2])
        line1.set_3d_properties(SetDataAnimation1[i][2])
    return line, line1


# 作静态图
ax.plot_wireframe(X, Y, Z, rstride=6, cstride=6, color="green")
ax.plot_surface(x, y1, z1, rstride=4, cstride=4, alpha=0.5, color="dodgerblue")
ax.plot_surface(x, y2, z2, rstride=4, cstride=4, alpha=0.5, color="yellow")


# 设置坐标轴参数
ax.set_title('圆柱体相交图示', fontproperties=font)
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_zlim(-1.2, 1.2)
ax.set_zlabel('Z轴', fontproperties=font)
ax.set_ylabel('Y轴', fontproperties=font)
ax.set_xlabel('X轴', fontproperties=font)


# 创建动画对象
ani = FuncAnimation(fig, UpdateData_Animation,range(1, 40),
                    interval=100,blit=False)

# 保存为GIF格式，注：需安装ImageMagick
# ani.save('D:\\PaintFigure_3D.gif',writer='imagemagick')
plt.show()
```
