```python
"""
 9.寻找“水仙花数”(三位自幂数)
"""

def self_power(n):
    '''
    求n位自幂数函数
    '''
    self_power_list = []
    for i in range(10**(n-1), 10**n - 1):
        self_power_check = 0
        single_num_list = single_num(i)
        for j in range(n):
            self_power_check += int(single_num_list[j])**n
        if self_power_check == i:
            self_power_list.append(self_power_check)
    return self_power_list

def single_num(num):
    num_str = str(num)
    single_num_list = []
    for x in num_str:
        single_num_list.append(x)
    return single_num_list

# 寻找“水仙花数”
self_power(3)
```

```python
"""
  10.寻找“完美数”
"""

import math

def perfect_num():
    '''
    求10000以内的完美数
    '''
    perfect_num_list = []
    for predict_num in range(1,10000):
        factor_sum = 0
        for factor in range(1, int(math.sqrt(predict_num) + 1)):
            if predict_num % factor == 0:
                factor_sum += factor
                if (factor > 1) and (predict_num / factor != factor):
                    factor_sum += predict_num / factor
        if factor_sum == predict_num:
            perfect_num_list.append(predict_num)
    return perfect_num_list
perfect_num()
```

```python
"""
 11.Craps赌博游戏

    玩家摇两颗色子.如果第一次摇出7点或11点.玩家胜;
    如果摇出2点、3点、12点,庄家胜;其他情况游戏继续;
    玩家再次要色子,如果摇出7点.庄家胜;如果摇出第一
    次摇的点数,玩家胜;否则游戏继续,玩家继续摇色子.
    玩家进入游戏时有1000元的赌注,全部输光游戏或达到5000结束
"""

from random import randint

def money_leave(money,debt):
    game_res = -1
    res = randint(1,12)
    if (res == 7) or (res == 11):
        game_res = 1
    elif (res == 2) or (res == 3) or (res == 12):
        game_res = 0
    else:
        while(game_res >= 0):
            res_new = randint(1,12)
            if res_new == 7:
                game_res = 0
            elif res_new == res:
                game_res = 1
            else:
                game_res = -1
    if game_res == 1:
        money += debt
        return money
    else:
        money -= debt
        return money
        
def main():
    money_init = 1000
    money = money_init
    while (0<money<5000):
        print('现有赌金：%d'% money)
        debt = int(input('请输入赌注：'))
        if (debt <= 0) and (debt > money):
            print('赌注不符合要求！')
            continue
        money = money_leave(money, debt)
    if money <= 0:
        print('您已破产！')
    else:
        print('您赌金已达到最大！')
“main()
```

```python
"""
 13.实现判断一个数是不是回文数的函数
"""

def is_palindrome(num):
    single_num_list = single_num(num)
    n = len(single_num_list)
    bool_check = True
    for i in range(n // 2):
        if single_num_list[i] != single_num_list[n - 1 - i]:
            bool_check = False
            break
    if bool_check:
        print('%d是回文数'%num)
    else:
        print('%d不是回文数'%num)

def single_num(num):
    num_str = str(num)
    single_num_list = []
    for x in num_str:
        single_num_list.append(x)
    return single_num_list
is_palindrome(12321)
```

```python
"""
 15.在屏幕上显示跑马灯文字
"""

import os 
import time

def fuc_1():
    content = '你最棒!'
    start = time.perf_counter ()
    end = start
    while end - start<3:
        # 清理屏幕上的输出
        os.system('cls')  # os.system('clear')
        print(content)
        # 休眠200毫秒
        time.sleep(0.2)
        content = content[1:] + content[0]
        end = time.perf_counter ()
fuc_1()
```

```python
"""
 16.产生指定长度的验证码（只由数字和字母构成）
"""

from random import randint

def generate_code(code_len=6):
    '''
    生成指定长度的验证码
    :param code_len: 验证码的长度（默认为6）
    :return: 指定长度只由数字和字母构成的验证码
    '''
    all_chars = '0123456789abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'
    code = ''
    for _ in range(code_len):
        index = randint(0,len(all_chars)-1)
        code += all_chars[index]
    return code
generate_code(8)
```

```python
"""
 17.打印杨辉三角
"""

def fuc2():
    row_num = int(input('请输入所要显示的行数：'))
    data = [[]]*row_num
    for row_ in range(len(data)):
        data[row_] = [None] * (row_ + 1) 
        for col_ in range(len(data[row_])):
            if (col_ == 0) or (col_ == row_):
                data[row_][col_] = 1
            else:
                data[row_][col_] = data[row_-1][col_-1] + data[row_-1][col_]
            print(data[row_][col_],end='')
        print()
fuc2()
    
```

```python
"""
 18.五子棋（9x9）游戏
"""

import os

def print_board(board):
    
    for row_list in board:
        row = ''
        for col in range(len(row_list)):
            row += row_list[col]
            if  len(row_list)-col > 1:
                row += '|'
        print(row)
        print('-+-+-+-+-+-+-+-+-')

def main():
    init_ = [[' ']*9 for i in range(9)]
    begin = True
    while begin:
        curr_ = init_
        begin = False
        counter = 0
        os.system('clear')
        print_board(curr_)
        turn_object = '己方'
        turn_perform = 'x'
        is_run = True
        while (is_run):
            move = input('轮到%s走棋, 请输入位置:(格式：x,y) ' % turn_object)
            x_1 = int(move.strip().split(',')[0])-1
            y_1 = int(move.strip().split(',')[1])-1
            if curr_[x_1][y_1] == ' ':
                counter += 1 
                curr_[x_1][y_1] = turn_perform
                if turn_object == '己方':
                    turn_perform = 'o'
                    turn_object = '对方'
                else:
                    turn_perform = 'x'
                    turn_object = '己方'
            else:
                print('此处已有，请重新走棋！')
             
            if judge(x_1 + 1,y_1 + 1, curr_):
                is_run = False
                print('{}已经输了！'.format(turn_object))
            os.system('clear')
            print_board(curr_)
        choice = input('再玩一局?(yes|no)')
        begin = choice == 'yes'
        
def judge(x,y,curr):
    x_0 = x
    y_0 = y
    curr_ = curr
    for direction in range(4):
        counter_r = 0
        counter_l = 0
        x_1 = x_0
        y_1 = y_0
        for i in range(4):
            [x_new , y_new] = location(x_1, y_1, direction, i,'right')
            if (x_new > len(curr_[0])) or (x_new < 1):
                break
            if (y_new > len(curr_)) or (y_new < 1):
                break
            if curr_[x_new-1][y_new-1] == curr_[x_0-1][y_0-1]:
                counter_r += 1
            else:
                break
        for j in range(4):
            [x_new ,y_new] = location(x_1, y_1, direction, j, 'left')
            if (x_new > len(curr_[0])) or (x_new < 1):
                break
            if (y_new > len(curr_)) or (y_new < 1):
                break
            if curr_[x_new-1][y_new-1]== curr_[x_0-1][y_0-1]:
                counter_l += 1
            else:
                break
        counter = counter_r + counter_l + 1
        print(counter)
        if counter == 5:
            judge_res = True
            break
        else:
            judge_res = False
    return judge_res

def location(x,y,a,b,s):
    location_new = []
    if s == 'right':
        if a == 0:
            location_new = [x,y-b-1]
        elif a == 1:
            location_new = [x+b+1,y-b-1]
        elif a == 2:
            location_new = [x+b+1,y]
        elif a == 3:
            location_new = [x+b+1,y+b+1]
    elif s == 'left':
        if a == 0:
            location_new = [x,y+b+1]
        elif a == 1:
            location_new = [x-b-1,y+b+1]
        elif a == 2:
            location_new = [x-b-1,y]
        elif a == 3:
            location_new = [x-b-1,y-b-1]
    return location_new

main()
```
