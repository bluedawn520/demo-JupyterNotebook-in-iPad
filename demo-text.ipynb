```python
"""
  10.寻找“完美数”
"""

import math

def perfect_num():
    '''
    求10000以内的完美数
    '''
    perfect_num_list = []
    for predict_num in range(1,10000):
        factor_sum = 0
        for factor in range(1, int(math.sqrt(predict_num) + 1)):
            if predict_num % factor == 0:
                factor_sum += factor
                if (factor > 1) and (predict_num / factor != factor):
                    factor_sum += predict_num / factor
        if factor_sum == predict_num:
            perfect_num_list.append(predict_num)
    return perfect_num_list
perfect_num()
```

```python
"""
 13.实现判断一个数是不是回文数的函数
"""

def is_palindrome(num):
    single_num_list = single_num(num)
    n = len(single_num_list)
    bool_check = True
    for i in range(n // 2):
        if single_num_list[i] != single_num_list[n - 1 - i]:
            bool_check = False
            break
    if bool_check:
        print('%d是回文数'%num)
    else:
        print('%d不是回文数'%num)

def single_num(num):
    num_str = str(num)
    single_num_list = []
    for x in num_str:
        single_num_list.append(x)
    return single_num_list
is_palindrome(12321)
```

```python
"""
 16.产生指定长度的验证码（只由数字和字母构成）
"""

from random import randint

def generate_code(code_len=6):
    '''
    生成指定长度的验证码
    :param code_len: 验证码的长度（默认为6）
    :return: 指定长度只由数字和字母构成的验证码
    '''
    all_chars = '0123456789abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'
    code = ''
    for _ in range(code_len):
        index = randint(0,len(all_chars)-1)
        code += all_chars[index]
    return code
generate_code(8)
```

```python
"""
 18.五子棋（9x9）游戏
"""

import os

def print_board(board):
    '''
    显示棋盘函数
    '''
    for row_list in board:
        row = ''
        for col in range(len(row_list)):
            row += row_list[col]
            if len(row_list) - col > 1:
                row += '|'
        print(row)
        print('-+-+-+-+-+-+-+-+-')

def judge(x, y, curr):
    '''
    判断是否当前位置已经是否可以结束游戏
    :param x: 横坐标
    :param y: 纵坐标
    :param curr: 存放数据的列表
    :return: True or False
    '''
    x_1 = x
    y_1 = y
    curr_ = curr
    for direction in range(4):
        counter_r = 0
        counter_l = 0
        for i in range(4):
            [x_new, y_new] = location(x_1, y_1, direction, i, 'right')
            if (x_new > len(curr_[0])) or (x_new < 1):
                break
            if (y_new > len(curr_)) or (y_new < 1):
                break
            if curr_[x_new - 1][y_new - 1] == curr_[x_0 - 1][y_0 - 1]:
                counter_r += 1
            else:
                break
        for j in range(4):
            [x_new, y_new] = location(x_1, y_1, direction, j, 'left')
            if (x_new > len(curr_[0])) or (x_new < 1):
                break
            if (y_new > len(curr_)) or (y_new < 1):
                break
            if curr_[x_new - 1][y_new - 1] == curr_[x_0 - 1][y_0 - 1]:
                counter_l += 1
            else:
                break
        counter = counter_r + counter_l + 1
        if counter == 5:
            judge_res = True
            break
        else:
            judge_res = False
    return judge_res

def location(x, y, a, b, s):
    '''
    获取临近的棋子位置函数
    :param x: 横坐标(int)
    :param y: 纵坐标(int)
    :param a: 方向1选择(int)
    :param b: 相邻棋子的个数
    :param s: 方向2选择('right' or 'left')
    :return: 临近棋子位置(list)
    '''
    location_new = []
    if s == 'right':
        if a == 0:
            location_new = [x, y - b - 1]
        elif a == 1:
            location_new = [x + b + 1, y - b - 1]
        elif a == 2:
            location_new = [x + b + 1, y]
        elif a == 3:
            location_new = [x + b + 1, y + b + 1]
    elif s == 'left':
        if a == 0:
            location_new = [x, y + b + 1]
        elif a == 1:
            location_new = [x - b - 1, y + b + 1]
        elif a == 2:
            location_new = [x - b - 1, y]
        elif a == 3:
            location_new = [x - b - 1, y - b - 1]
    return location_new

def gobang():
    '''
    五子棋游戏（9x9）
    '''
    init_ = [[' '] * 9 for i in range(9)]
    begin = True
    while begin:
        curr_ = init_
        begin = False
        counter = 0
        # os.system('clear')
        print_board(curr_)
        turn_object = '己方'
        turn_perform = 'x'
        is_run = True
        while (is_run):
            print('输入：STOP，中止当前进行的游戏！')
            move = input('轮到%s走棋, 请输入位置(格式：4,4):' % turn_object)
            if move == 'STOP':
                break
            else:
                if move.strip().split(',')[0].isalnum() and move.strip().split(',')[1].isalnum():
                    x_1 = int(move.strip().split(',')[0]) - 1
                    y_1 = int(move.strip().split(',')[1]) - 1
                    if (len(curr_[0]) > x_1 >= 0) and (len(curr_) > y_1 >= 0):
                        if curr_[x_1][y_1] == ' ':
                            counter += 1
                            curr_[x_1][y_1] = turn_perform
                            if turn_object == '己方':
                                turn_perform = 'o'
                                turn_object = '对方'
                            else:
                                turn_perform = 'x'
                                turn_object = '己方'
                        else:
                            print('此处已有，请重新走棋！')

                        if judge(x_1 + 1, y_1 + 1, curr_):
                            is_run = False
                            print('{}已经输了！'.format(turn_object))
                        # os.system('clear')
                        print_board(curr_)
                    else:
                        print('输入位置超出棋盘！')
                else:
                    print('输入格式不正确！')
        choice = input('再玩一局?(yes|no)')
        begin = choice == 'yes'

gobang()
```

```python
'''
 19.数字时钟
'''

import time

class Clock(object):
    
    # Python中的函数是没有重载的概念的
    # 因为Python中函数的参数没有类型而且支持缺省参数和可变参数
    # 用关键字参数让构造器可以传入任意多个参数来实现其他语言中的构造器重载
    def __init__(self,**kw):
        '''
        初始化方法
        '''
        if ('hour' in kw) and ('minute' in kw) and ('second' in kw):
            self._h = kw['hour']
            self._m = kw['minute']
            self._s = kw['second']
        else:
            tm =time.localtime(time.time())
            self._h = tm.tm_hour
            self._m = tm.tm_min
            self._s = tm.tm_sec
            
    def run(self):
        self._s += 1
        if self._s == 60:
            self._s = 0
            self._m +=1
            if self._m == 60:
                self._m = 0
                self._h +=1
                if self._h == 24:
                    self._h = 0
    def style(self,obj):
        if len(str(obj)) == 1:
            style_new = '0' + str(obj)
        else:
            style_new = str(obj)
        return style_new
    def show(self):
        style_h = self.style(self._h)
        style_m = self.style(self._m)
        style_s = self.style(self._s)
        show_text = '{}:{}:{}'.format(style_h, style_m, style_s)
        return show_text
        
def main():
    clo =Clock()
    index = 0
    while (index < 10):
        index += 1
        print(clo.show())
        time.sleep(1)
        clo.run()
        
main()
```

```python
'''
 20.移动平面上的点到指定位置并画出路径
'''
import math
import turtle 

class Point(object):
    
    def __init__(self,x=0,y=0):
        self._x = x
        self._y = y
    
    def move_to(self,x,y):
        self._x = x
        self._y = y
        
    def move_pass(self,dx,dy):
        self._x += dx
        self._y += dy
        
    def trace_paint(self,new):
        turtle.screensize()
        turtle.speed(5)
        turtle.penup()
        turtle.goto(self._x, self._y)
        turtle.pendown()
        turtle.dot(10, 'red')
        turtle.pencolor('green')
        turtle.pensize(2)
        turtle.goto(new._x, new._y)
        turtle.dot(10, 'blue')
        turtle.hideturtle()
        

poi = Point(10,20)
poi_1 = Point()
poi_1.move_to(-30,50)
poi.trace_paint(poi_1)
```

```python
#  参考网址：  https://www.jb51.net/article/134148.htm
#    使用@property包装器来包装getter和setter方法
'''
 21.修一个游泳池,半径(以米为单位)在程序运行时输入,游泳池外修一条3米宽的过道,
过道的外侧修一圈围墙;已知过道的造价为25元每平米,围墙的造价为32.5元每米,
输出围墙和过道的总造价分别是多少钱(精确到小数点后2位)
'''

import math

class swimming_pool(object):
    
    def __init__(self,radius):
        self._radius = radius
        
    @property
    def radius(self):
        return self._radius
    
    @radius.setter
    def radius(self,radius):
        self._radius = radius if radius >0 else 0
    
    @property
    def perimeter(self):
        return 2 * math.pi * self._radius
    
    @property
    def area(self):
        return math.pi * self._radius * self._radius9

def main():
    radius = float(input('请输入游泳池的半径(米): '))
    small_circle = swimming_pool(radius)
    big_circle = swimming_pool(radius + 3)
    print('围墙的造价为: ￥%.1f元' % (big_circle.perimeter * 115))
    print('过道的造价为: ￥%.1f元' % ((big_circle.area - small_circle.area) * 65))
 
    
main()
```

```python
'''  
  23. __slots__魔法 :  需要限定自定义类型的对象只能绑定某些属性，可以通过在类中
  定义__slots__变量来进行限定。需要注意的是__slots__的限定只对当前类的对象
  生效，对子类并不起任何作用
'''
#  参考网址： http://m.biancheng.net/view/2291.html
#  参考网址:  https://www.jb51.net/article/120232.html

def cat_walk(self):
    print('%s漫走在世界之巅！'%self._name)

cat_fly = lambda self: print('%s自由的飞翔！'%self._name)

class Cat:
    def __init__(self,name):
        self._name = name
        
cat_0 = Cat('Tom')
cat_1 = Cat('Jim')
# Cat动态添加了walk()方法，为类动态添加方法时不需要使用 MethodType 
# 进行包装，该函数的第一个参数会自动绑定
Cat.walk = cat_walk
Cat.fly = cat_fly

cat_0.walk()
cat_1.walk()
cat_0.fly()


#  __slots__限制为某个类动态添加属性和方法(并不限制通过类来动态添加属性或方法)
class Dog:
    __slots__ = ('walk', 'age', '_name')
    def __init__(self, name):
        self._name = name
    pass
        
dog_0 = Dog('Jerk')
dog_1 = Dog('Peppy')
from types import MethodType
# 只允许动态为实例添加walk、age、name这2个属性或方法
dog_0.walk = MethodType(lambda self: print('%s在游走！'%self._name), dog_0)
dog_0.age = 18

dog_0.walk()
print(dog_0.age)

Dog.speed = lambda self: print('%s速度很快！'%self._name)
dog_0.speed()

class DogOffpring(Dog):
    def __init__(self,name):
        super().__init__(name)
    pass
dogoffspring_0 = DogOffpring('Jerk-son')

dogoffspring_0.rank = 1
print('%s等级为%d'%(dogoffspring_0._name, dogoffspring_0.rank))
```

```python
'''
 24.静态方法
     类中定义的方法既可以是对象方法，也可以不是对象方法（如下）
'''

import math 

class Triangle(object):
    
    def __init__(self, a, b, c):
        self._a = a
        self._b = b
        self._c = c
    @staticmethod
    def is_valid(a, b, c):
        return (a + b > c) and (a + c > b) and (c + b > a)
    @property
    def perimeter(self):
        return self._a + self._b + self._c
    @property
    def area(self):
        half = self.perimeter/2
        return math.sqrt(half * (half - self._a) *
                    (half - self._b) * (half - self._c))
        
def main():
    a = int(input('请输入三角形边长1：'))
    b = int(input('请输入三角形边长2：'))
    c = int(input('请输入三角形边长3：'))
    if Triangle.is_valid(a, b, c):
        t = Triangle(a, b, c)
        print('周长： ',t.perimeter)
        print('面积： ',t.area)
    else:
        print('无法构成三角形！')
        
main()
```

```python
'''
 25.定义类方法，
 第一个参数名约定为cls,代表当前类相关信息的对象/类的元数据对象（类本身也是一个对象）
'''

import time

class Clock(object):
    def __init__(self, h, m , s):
        self._h = h
        self._m = m
        self._s = s
    
    @classmethod
    def now(cls):
        ctime = time.localtime(time.time())
        return cls(ctime.tm_hour, ctime.tm_min, ctime.tm_sec)
    
    def run(self):
        self._s += 1
        if self._s == 60:
            self._s = 0
            self._m +=1
            if self._m == 60:
                self._m = 0
                self._h +=1
                if self._h == 24:
                    self._h = 0
    def style(self,obj):
        if len(str(obj)) == 1:
            style_new = '0' + str(obj)
        else:
            style_new = str(obj)
        return style_new
    def show(self):
        style_h = self.style(self._h)
        style_m = self.style(self._m)
        style_s = self.style(self._s)
        show_text = '{}:{}:{}'.format(style_h, style_m, style_s)
        return show_text
        
def main():
    clo = Clock.now()
    i = 0
    while i < 50:
        i += 1
        print(clo.show())
        time.sleep(1)
        clo.run()

main()
```

```python
'''
 26.类之间的关系；多态
     is-a继承/泛化, has-a关联, use-a依赖
'''
'''
在实际开发中，我们经常会用子类对象去替换掉一个父类对象，
这是面向对象编程中一个常见的行为，对应的原则称之为里氏替换原则 
'''

class Person(object):
    def __init__(self,name,sex,age):
        self._name = name
        self._sex = sex
        self._age = age
    @property
    def name(self):
        return self._name
    @property
    def sex(self):
        return self._sex
    @property
    def age(self):
        return self._age
    @age.setter
    def age(self, age):
        self._age = age
    def explain(self):
        print('%s is a person.'%self._name)

class Student(Person):
    
    def __init__(self,name,sex,age,grade):
        super().__init__(name,sex,age)
        self._grade = grade
    @property
    def grade(self):
        return self._grade
    @grade.setter
    def grade(self,grade):
        self._grade = grade
    def explain1(self):
        print('%s is a student.'%self._name)
    
        
class Teacher(Person):
    
    def __init__(self, name, age, sex, title):
        super().__init__(name, age, sex)
        self._title = title
    @property
    def title(self):
        return self._title
    @title.setter
    def title(self, title):
        self._title = title
    def explain1(self):
        print('%s is a teacher.'%self._name)
    
def main():
    stu_0 = Student('bluedawn','male',21,'大三')
    tea_0 = Teacher('william','male',33,'python')
    stu_0.explain()
    tea_0.explain()
    
main()
```

```python
'''
 27.创建抽象类（即不能够创建对象的类）
 Python从语法层面并没有像Java或C#那样提供对抽象类的支持，但是我们可以
 通过abc模块的ABCMeta元类和abstractmethod包装器来达到抽象类的效果，如
 果一个类中存在抽象方法那么这个类就不能够实例化（创建对象）
'''

from abc import ABCMeta, abstractmethod

class Pet(object,metaclass=ABCMeta):
    
    def __init__(self,name):
        self._name = name
    @abstractmethod
    def make_voice(self):
        pass
class Dog(Pet):
    def make_voice(self):
        print('%s: 汪汪汪...' % self._name)
class Cat(Pet):
    def make_voice(self):
        print('%s: 喵...喵...' % self._name)

def main():
    pets = [Dog('AA'), Cat('BB'), Dog('CC')]
    for pet in pets:
        pet.make_voice()
        
main()
```
